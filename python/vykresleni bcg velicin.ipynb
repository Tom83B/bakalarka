```python
import EOS
import derivatives as d
from scipy.interpolate import UnivariateSpline

rG = np.empty(z.size)
rG_ext = np.empty(z.size)

for i in range(0,z.size):
    rG[i] = EOS.radG(T[i], P[i], z[i])
    rG_ext[i] = EOS.radG(ext.T(z[i]), ext.p(z[i]), z[i])

convG = np.empty(z.size)
convG_ext = np.empty(z.size)

for i in range(0,z.size):
#    convG[i] = d.TDeriv(T[i], P[i], z[i])
    convG_ext[i] = d.TDeriv(ext.T(z[i]), ext.p(z[i]), z[i])

G = np.empty(z.size)
G_ext = np.empty(z.size)

lnP = UnivariateSpline(z, np.log(P), k=3, s=0)
dlnP = lnP.derivative()
lnP_ext = UnivariateSpline(z, np.log(ext.p(z)), k=3, s=0)
dlnP_ext = lnP_ext.derivative()
lnT = UnivariateSpline(z, np.log(T), k=3, s=0)
dlnT = lnT.derivative()
lnT_ext = UnivariateSpline(z, np.log(ext.T(z)), k=3, s=0)
dlnT_ext = lnT_ext.derivative()

G = lambda z: dlnT(z)/dlnP(z)
G_ext = lambda z: dlnT_ext(z)/dlnP_ext(z)

fig = plt.figure(figsize=(12,6), dpi=300)

axes = fig.add_axes([0,0,1,1])

m=0
n=500
#axes.plot(z[m:n], adG[m:n], c='blue', label = 'adG')
#axes.plot(z[m:n], adG_ext[m:n], c='blue', linestyle = 'dashed')
#axes.plot(z[m:n], ext.adG(z[m:n]), c='red', linestyle = '-.')
#axes.plot(z[m:n], rG[m:n], c='red', label = 'radG')
#axes.plot(z[m:n], rG_ext[m:n], c='red', linestyle = 'dashed')
#axes.plot(z[m:n], convG[m:n], c='green', label = 'convG')
#axes.plot(z[m:n], convG_ext[m:n], c='green', linestyle = 'dashed')
axes.plot(z[m:n], G(z[m:n]), c='black', label = 'd(lnT)/d(lnP)')
axes.plot(z[m:n], G_ext(z[m:n]), c='black', linestyle = 'dashed')
#axes.plot(z, ext.r(z), c='black', linestyle='dashed')
axes.set_xlabel("$z\ (\mathrm{m})$", fontsize=18)
#axes.set_ylabel(r"$ \nabla_\mathrm{ad} $",fontsize=18)
#axes.legend()

fig.savefig("gradienty.png",bbox_inches='tight')
```
