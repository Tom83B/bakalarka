```python
frad, fconv, extfrad, extfconv = get_flows()
```

```python
fig = plt.figure(figsize=(12,6), dpi=300)

axes = fig.add_axes([0,0,1,1])

axes.plot(z,-frad, c='b', label='$F_{\mathrm{rad}}$')
axes.plot(z,-fconv, c='g', label='$F_{\mathrm{conv}}$')
axes.plot(z,-extfrad, c='b', linestyle='dashed')
axes.plot(z,-extfconv, c='g', linestyle='dashed')
axes.plot(z,-frad-fconv, c='r', label='$F_{\mathrm{sum}}$')
axes.plot(z,-extfrad-extfconv, c='r', linestyle='dashed')
axes.set_xlabel("$z\ (\mathrm{m})$", fontsize=18)
axes.set_ylabel("$ F\ (\mathrm{W/m^2}) $",fontsize=18)
axes.set_yscale('log')
axes.legend(prop={'size':18})

fig.savefig("flows.png",bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(12,6), dpi=300)

axes = fig.add_axes([0,0,1,1])

axes.plot(z,-frad-fconv, c='black')
axes.plot(z,-extfrad-extfconv, c='black', linestyle='dashed')
axes.set_xlabel("$z\ (\mathrm{m})$", fontsize=18)
axes.set_ylabel("$ F\ (\mathrm{W/m^2}) $",fontsize=18)
axes.set_yscale('log')

fig.savefig("flowsum.png",bbox_inches='tight')
```
